# Credit-Shield: Evaluación de Sentimiento
Este notebook simula la evaluación de frases bancarias y visualiza la confianza del modelo DistilBERT.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulación de resultados de evaluación
data = {
    'phrase': [
        'Excellent service', 
        'High interest rates', 
        'Easy to use app', 
        'Card lost, help!', 
        'I love the cashback'
    ],
    'sentiment': ['POSITIVE', 'NEGATIVE', 'POSITIVE', 'NEGATIVE', 'POSITIVE'],
    'confidence': [0.99, 0.85, 0.98, 0.70, 0.99]
}

df = pd.DataFrame(data)
df.head()
```

```python
# Graficando la confianza del modelo
plt.figure(figsize=(10, 6))
colors = ['green' if s == 'POSITIVE' else 'red' for s in df['sentiment']]
plt.bar(df['phrase'], df['confidence'], color=colors)
plt.title('Confianza del Modelo Credit-Shield por Frase')
plt.xlabel('Frase de Cliente')
plt.ylabel('Score de Confianza')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
